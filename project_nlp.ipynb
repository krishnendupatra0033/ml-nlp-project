```python
import fitz  # PyMuPDF
import spacy

# Define the function with a proper parameter name, e.g., 'pdf_path'
def extract_text_from_pdf(pdf_path):
    # Use the parameter 'pdf_path' to open the file
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Load the English NER model
nlp = spacy.load("en_core_web_sm")

def apply_ner(text, chunk_size=500000): # Added chunk_size parameter
    # Split the text into chunks
    text_chunks = [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]

    print("Processing text in chunks...")
    for i, chunk in enumerate(text_chunks):
        print(f"Processing chunk {i+1}/{len(text_chunks)}")
        doc = nlp(chunk)
        # Process named entities for the current chunk
        for ent in doc.ents:
            print(f"{ent.text} ({ent.label_}): {spacy.explain(ent.label_)}")

fghfgh = extract_text_from_pdf("/content/john-stuart-mill_system-of-logic.pdf")
apply_ner(fghfgh) # Call the function with the extracted text
```
